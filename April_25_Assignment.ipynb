```python
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ubizareie/python-DA/main/StudentsPerformance.csv.xls') # My dataset

#plt.plot(df [['reading score','math score']], marker = 'p', c= "b")
plt.plot(df['reading score'], marker = 'p', c= "r", ls ='-', lw ='.5', label = 'Reading')
#plt.legend()
plt.title ('Student Performance in READING', c= 'k', weight= 'bold')
plt.xlabel('FREQUENCY')
plt.ylabel('SCORE')
plt.grid()
plt.show()
plt.plot(df['math score'], marker = '>', c = '#6932a8', ls = 'dotted', lw ='.5')
plt.title ('Student Performance in MATH', c= 'k', weight= 'demibold')
plt.xlabel('FREQUENCY')
plt.ylabel('SCORE')
plt.show()
plt.plot(df['writing score'], marker = 7 , c = '#329aa8' , ls = 'dashdot', lw= 1.7 )
plt.title ('Student Performance in WRITING', c= 'k', weight= 'bold')
plt.xlabel('FREQUENCY')
plt.ylabel('SCORE')
plt.show()
plt.plot(df[['reading score', 'math score', 'writing score']])
plt.title ('Comparing Student Performance in MATH and ELA(reading-writing)', c= 'k', weight= 'extra bold')
plt.xlabel('FREQUENCY')
plt.ylabel('SCORE')
plt.show()
df.plot.scatter (x = 'reading score', y = 'math score')
df.plot.scatter (x = 'math score', y = 'reading score')
df.plot.scatter (x = 'reading score', y = 'writing score')
plt.show()
```
